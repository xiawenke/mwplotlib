```python
from mwplotlib import MilkyWay # import mwplotlib package
import matplotlib.pyplot as plt # import matplotlib.pyplot
import astropy.units as u # import astropy.units
```

```python
# Add the mwplotlib to the matplotlib.pyplot
plt = MilkyWay.mw_plt(plt) # Add the mwplotlib to the matplotlib.pyplot

# This plt can still be used as the original plt
plt.figure(figsize=(7, 4)) # Set the figure size
plt.title("Example") # Set the title
plt.xlabel("X") # Set the x label
plt.ylabel("Y") # Set the y label
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) # Plot the data
```

```python
# Plot the Milky Way in polar coordinates (region-style)
plt.figure(figsize=(9, 9)) # Set the figure size
plt.MilkyWay.shape(plot_type="polar", style="region") # Plot the Milky Way in polar coordinates (region-style)
plt.legend() # Show the legend

# Plot the Milky Way in polar coordinates (line-style)
plt.figure(figsize=(9, 9)) # Set the figure size
plt.MilkyWay.shape(plot_type="polar", style="line") # Plot the Milky Way in polar coordinates (line-style)
plt.legend() # Show the legend
```

```python
# Plot the Milky Way in Cartesian coordinates (region-style)
plt.figure(figsize=(9, 9)) # Set the figure size
plt.MilkyWay.shape(plot_type="cartesian", style="region") # Plot the Milky Way in Cartesian coordinates (region-style)
plt.legend() # Show the legend

# Plot the Milky Way in Cartesian coordinates (line-style)
plt.figure(figsize=(9, 9)) # Set the figure size
plt.MilkyWay.shape(plot_type="cartesian", style="line") # Plot the Milky Way in Cartesian coordinates (line-style)
plt.legend() # Show the legend
```

```python
# You can add more arguments to the shape() function. 
# Those arguments will be passed to the matplotlib.pyplot.plot() function.
# For example, you can change the linewidth (lw) of the shape.
plt.figure(figsize=(9, 9)) # Set the figure size
plt.MilkyWay.shape(plot_type="cartesian", style="region", lw=1,
                   legend=False # Hide the legend
                   ) # Plot the Milky Way in Cartesian coordinates (region-style)

# You can add other elements to the plot.
plt.plot(8.3, 0, "*", color="red") # Plot the Sun
plt.text(8.45, 0.15, "Sun") # Add text to the plot (Sun)
plt.text(2.6, 2.5, "Norma") # Add text to the plot (Norma)


# You can further customize the plot.
plt.grid() # Add grid
plt.xlabel("X (kpc)") # Set the x label
plt.ylabel("Y (kpc)") # Set the y label

plt.legend() # Show the legend, but since we set legend=False in the shape() function, the legend will be empty.
```

```python
# The unit of the plot is in kpc by default.
# You can change the unit by passing the unit argument to the shape() function.
# For example, you can change the unit to pc. 
# (All the astropy.units are supported. See https://docs.astropy.org/en/stable/units/ for more information.)
plt.figure(figsize=(9, 9)) # Set the figure size
plt.MilkyWay.shape(plot_type="cartesian", style="region", lw=1, unit="pc") # Plot the Milky Way in Cartesian coordinates (region-style)
# plt.MilkyWay.shape(plot_type="cartesian", style="region", lw=1, unit=u.pc) # Using u.pc instead of "pc" is also supported. (version >= 1.1)

# You can further customize the plot.
plt.grid() # Add grid
plt.xlabel("X (pc)") # Set the x label
plt.ylabel("Y (pc)") # Set the y label

plt.legend() # Show the legend
```

```python
# To higher/lower the resolution of the plot, you can pass the step argument to the shape() function.
# For example, you can change the step to 10 kpc.
# Be careful that the unit of the step argument is always in kpc. 
plt.figure(figsize=(9, 9)) # Set the figure size
plt.MilkyWay.shape(plot_type="cartesian", style="region", lw=1, unit="kpc", step=10) # Plot the Milky Way in Cartesian coordinates (region-style)

# You can further customize the plot.
plt.grid() # Add grid
plt.xlabel("X (kpc)") # Set the x label
plt.ylabel("Y (kpc)") # Set the y label

plt.legend() # Show the legend
```

```python
# In addition to passing arguments to the shape() function that applies to all the arms,
# you can also use config argument to pass arguments to specific arms.
# For example, you can highlight the Local arm by changing the color and linewidth of it.
plt.figure(figsize=(9, 9)) # Set the figure size
plt.MilkyWay.shape(plot_type="cartesian", style="region", unit="kpc",
                        config={
                            "Local": {"color": "red", "lw": 2},
                        }, # Config that applies to the Local arm only
                        color="black", lw=0.5 # Config that applies to all the (other) arms
                   ) # Plot the Milky Way in Cartesian coordinates (region-style)

# You can further customize the plot.
plt.grid() # Add grid
plt.xlabel("X (kpc)") # Set the x label
plt.ylabel("Y (kpc)") # Set the y label

plt.legend() # Show the legend
```

```python
# You can also plot using the ax object. This allows you to plot multiple plots in one figure.
# Here is an example.

fig = plt.figure(figsize=(15, 15)) # Create a figure

# First plot - Cartesian (region-style)
ax1 = plt.subplot(221) # Create a subplot
ax1 = MilkyWay.mw_plt(ax1) # Add the mwplotlib to the subplot
ax1.set_title("Cartesian (region-style)") # Set the title of the subplot
ax1.set_xlabel("X (kpc)") # Set the x-label of the subplot
ax1.set_ylabel("Y (kpc)") # Set the y-label of the subplot
ax1.MilkyWay.shape(plot_type="cartesian", style="region", lw=1) # Plot the Milky Way
ax1.legend() # Show the legend

# Second plot - Cartesian (line-style)
ax2 = plt.subplot(222) # Create a subplot
ax2 = MilkyWay.mw_plt(ax2) # Add the mwplotlib to the subplot
ax2.set_title("Cartesian (line-style)") # Set the title of the subplot
ax2.set_xlabel("X (kpc)") # Set the x-label of the subplot
ax2.set_ylabel("Y (kpc)") # Set the y-label of the subplot
ax2.MilkyWay.shape(plot_type="cartesian", style="line", lw=1) # Plot the Milky Way
ax2.legend() # Show the legend

# Third plot - Polar (region-style)
ax3 = plt.subplot(223, projection='polar') # Create a subplot, IT IS IMPORTANT TO SPECIFY THE PROJECTION
ax3 = MilkyWay.mw_plt(ax3) # Add the mwplotlib to the subplot
ax3.set_title("Polar (region-style)") # Set the title of the subplot
ax3.set_xlabel("R (kpc)") # Set the x-label of the subplot
ax3.set_ylabel("Theta (deg)") # Set the y-label of the subplot
ax3.MilkyWay.shape(plot_type="polar", style="region", lw=1) # Plot the Milky Way
ax3.legend() # Show the legend

# Fourth plot - Polar (line-style)
ax4 = plt.subplot(224, projection='polar') # Create a subplot, IT IS IMPORTANT TO SPECIFY THE PROJECTION
ax4 = MilkyWay.mw_plt(ax4) # Add the mwplotlib to the subplot
ax4.set_title("Polar (line-style)") # Set the title of the subplot
ax4.set_xlabel("R (kpc)") # Set the x-label of the subplot
ax4.set_ylabel("Theta (deg)") # Set the y-label of the subplot
ax4.MilkyWay.shape(plot_type="polar", style="line", lw=1) # Plot the Milky Way
ax4.legend() # Show the legend
```
