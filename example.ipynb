```python
from mwplotlib import MilkyWay
import matplotlib.pyplot as plt
```

```python
# Add the mwplotlib to the matplotlib.pyplot
plt = MilkyWay.mw_plt(plt) 

# This plt can still be used as the original plt
plt.figure(figsize=(7, 4))
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

```python
# Plot the Milky Way in polar coordinates (region-style)
plt.figure(figsize=(9, 9))
plt.MilkyWay.shape(plot_type="polar", style="region")

# Plot the Milky Way in polar coordinates (line-style)
plt.figure(figsize=(9, 9))
plt.MilkyWay.shape(plot_type="polar", style="line")
```

```python
# Plot the Milky Way in Cartesian coordinates (region-style)
plt.figure(figsize=(9, 9))
plt.MilkyWay.shape(plot_type="cartesian", style="region")

# Plot the Milky Way in Cartesian coordinates (line-style)
plt.figure(figsize=(9, 9))
plt.MilkyWay.shape(plot_type="cartesian", style="line")
```

```python
# You can add more arguments to the shape() function. 
# Those arguments will be passed to the matplotlib.pyplot.plot() function.
# For example, you can change the linewidth (lw) of the shape.
plt.figure(figsize=(9, 9))
plt.MilkyWay.shape(plot_type="cartesian", style="region", lw=1)

# You can add other elements to the plot.
plt.plot(8.3, 0, "*", color="red")
plt.text(8.45, 0.15, "Sun")
plt.text(2.6, 2.5, "Norma")

# You can further customize the plot.
plt.grid()
plt.xlabel("X (kpc)")
plt.ylabel("Y (kpc)")
```

```python
# The unit of the plot is in kpc by default.
# You can change the unit by passing the unit argument to the shape() function.
# For example, you can change the unit to pc. 
# (There are other units supported, such as m, km, lightyear, etc.)
plt.figure(figsize=(9, 9))
plt.MilkyWay.shape(plot_type="cartesian", style="region", lw=1, unit="pc")

# You can further customize the plot.
plt.grid()
plt.xlabel("X (pc)")
plt.ylabel("Y (pc)")
```

```python
# To higher/lower the resolution of the plot, you can pass the step argument to the shape() function.
# For example, you can change the step to 10 kpc.
# Be careful that the unit of the step argument is always in kpc. 
plt.figure(figsize=(9, 9))
plt.MilkyWay.shape(plot_type="cartesian", style="region", lw=1, unit="kpc", step=10)

# You can further customize the plot.
plt.grid()
plt.xlabel("X (kpc)")
plt.ylabel("Y (kpc)")
```

```python
# Plot in subplots
fig, axs = plt.subplots(2, 2, figsize=(9, 9)) # Create a 2x2 subplots

# First plot
axs[0, 0] = MilkyWay.mw_plt(axs[0, 0]) # Add the mwplotlib to the matplotlib.pyplot
axs[0, 0].set_title("Cartesian (region)") # Set the title
axs[0, 0].MilkyWay.shape(plot_type="cartesian", style="region", lw=1, unit="pc") # Plot the Milky Way, Cartesian (region)
axs[0, 0].grid() # Add grid
axs[0, 0].set_xlabel("X (pc)") # Set the label of the x-axis
axs[0, 0].set_ylabel("Y (pc)") # Set the label of the y-axis

# Second plot
axs[0, 1] = MilkyWay.mw_plt(axs[0, 1]) # Add the mwplotlib to the matplotlib.pyplot
axs[0, 1].set_title("Cartesian (line)") # Set the title
axs[0, 1].MilkyWay.shape(plot_type="cartesian", style="line", lw=1, unit="pc") # Plot the Milky Way, Cartesian (line)
axs[0, 1].grid() # Add grid
axs[0, 1].set_xlabel("X (pc)") # Set the label of the x-axis
axs[0, 1].set_ylabel("Y (pc)") # Set the label of the y-axis

# Third plot
```
